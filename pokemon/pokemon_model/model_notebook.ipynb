```python
import pandas as pd
import numpy as np
pokemon = pd.read_csv("pokemon.csv")
```

```python
df = pd.DataFrame()
df[["Type 1", "Type 2", "Attack", "Defense", "Speed", "Name"]]= pokemon.loc[:, ["Type 1", "Type 2", "Attack", "Defense", "Speed", "Name"]]
```

```python
df
```

```python
#replacing np.nan with 'No Secondary Type'
df["Type 2"] = df["Type 2"].replace(np.nan, "No Secondary Type")
```

```python
df
```

```python
#Scaling Attack, Defense, Speed
from sklearn.preprocessing import MinMaxScaler
mm = MinMaxScaler()
df.iloc[:, 2:-1] = mm.fit_transform(df.iloc[:, 2:-1])
df.iloc[:,2:-1] = df.iloc[:,2:-1].applymap(lambda x:x*100)
```

```python
#creating fake data...
import random
import math
def temp_func(x):
    opt = random.choice([0,1,2,3])
    if opt == 0:
        #floor operation
        x = math.floor(x)
    elif opt == 1:
        #ceil operation
        x = math.ceil(x)
    elif opt == 2:
        #add some value
        x = x + random.choice([1,2,3,4,5,6])
    else:
        #subtract some value
        x = x - random.choice([1,2,3,4,5,6])
    return x

temp_df = df.copy()
temp_df.iloc[:,2:-1] = temp_df.iloc[:,2:-1].applymap(temp_func)

temp_df2 = df.copy()
temp_df2.iloc[:,2:-1] = temp_df2.iloc[:,2:-1].applymap(temp_func)

temp_df3 = df.copy()
temp_df3.iloc[:,2:-1] = temp_df3.iloc[:,2:-1].applymap(temp_func)

temp_df4 = df.copy()
temp_df4.iloc[:,2:-1] = temp_df4.iloc[:,2:-1].applymap(temp_func)
```

```python
df = pd.concat([df, temp_df, temp_df2, temp_df3])
df
```

```python
#X -> Independent variables
#y -> Dependent variable
X = df.iloc[:,:-1].values
y = df.iloc[:,-1].values
```

```python
#Encoding Categorical data: Type 1, Type 2
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers = [('encoder1', OneHotEncoder(sparse = False ), [0,1])], remainder = "passthrough" )
X = np.array(ct.fit_transform(X))
```

```python
#Encoding Categorical data: Name
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
le.fit(y)
y = le.transform(y)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier.fit(X, y)
```
